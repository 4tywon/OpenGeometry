```python
# Imports for Graphics
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
from matplotlib import pylab
```

```python
# Object managing the manipulation of the Polygon
class AntTrail:
    def __init__(self, points):
        self.size = len(points)
        self.cur = points
        self.itr = 0
        self.display()
    def update(self):
        def midpoints(a,b):
            return ((a[0]+b[0])/2, (a[1] + b[1]) / 2)
        arr = []
        for i in range(len(self.cur)):
            arr.append(midpoints(self.cur[i], self.cur[(i + 1) % self.size]))
        self.cur = arr
        self.itr = self.itr + 1
    def scale(self):
        arr = self.cur
        mx = max(arr, key = lambda x: x[0])[0]
        mnx = min(arr, key = lambda x: x[0])[0]
        my = max(arr, key = lambda x: x[1])[1]
        mny = min(arr, key = lambda x: x[1])[1]
        return (mx, my, mnx, mny)
            
    def display(self):
        codes = [Path.MOVETO] + (self.size - 1)* [Path.LINETO] + [Path.CLOSEPOLY]
        verts = self.cur + [self.cur[0]]
        path = Path(verts, codes)
        fig = plt.figure()
        s = self.scale()
        deltax = 0.1*(s[0] - s[2])
        deltay = 0.1*(s[1] - s[3])
        plt.axis([s[2] - deltax, s[0] + deltax, s[3] - deltay, s[1] + deltay], label = str(self.itr))
        ax = fig.add_subplot(111)
        patch = patches.PathPatch(path, facecolor='none', lw=2)
        ax.add_patch(patch)
        plt.show()

    def step(self):
        self.update()
        self.display()
```

```python
# Run a simulation of n steps on a polygon represented by a sequential list of vertices
def simulate(vertices, steps):
    print("--------------------------")
    print("Original Image: ")
    print("--------------------------")
    x = AntTrail(vertices)
    for i in range(steps):
        print("--------------------------")
        print("Step Number "+ str(x.itr + 1))
        print("--------------------------")
        x.step()
```

```python
# Test Case
simulate([(0,-1), (2,2), (0.2, 0.8), (-10, -8), (1, 0),(0,1),(-4, 20), (-4, -20), (3,4), (5,2)], 40)
```
